# Pattern Problems in Python
Practice problems to strengthen loop logic and nested iteration.

```python
## Rectangular Star Pattern

def pattern1(num):
    for i in range(num):
        for j in range(num):
            print("*", end=" ")
        print()

n = int(input("Enter a number:"))
pattern1(n)
```

```python
## Right-Angled Triangle Star Pattern

def pattern2(num):
    for i in range(num):
        for j in range(i+1):
            print("*", end=" ")
        print()

n = int(input("Enter a number: "))
pattern2(n)
```

```python
## Right-Angled Number Pyramid Pattern


def pattern3(num):
    for i in range(num):
        for j in range(i+1):
            print(j+1, end=" ")
        print()


n = int(input("Enter a number: "))
pattern3(n)
```

```python
## Right-Angled Number Pyramid II Pattern


def pattern4(num):
    for i in range(num):
        for j in range(i+1):
            print(i+1, end=" ")
        print()


n = int(input("Enter a number: "))
pattern4(n)
```
