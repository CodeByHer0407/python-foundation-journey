```python
## Iterating through a list

fruits = ["apple", "banana", "strawberry", "raspberry", "grapes"]
for fruit in fruits:
    print(fruit)
    
```

```python
## Iterating through a list using index

for index,fruit in enumerate(fruits):
    print(f"Index: {index}, Fruit {index}: {fruit}")
```

```python
## List comprehension

lst2 = []
for x in range(10):
    lst2.append(x**2)

print(lst2)
```

```python
## [expression for item in iterable]

squared_lst = [x**2 for x in range(10)]
print(squared_lst)
```

##### List Comprehension

Basic Syntax            [expression for item in iterable]

with conditional logic    [expression for item in iterable if condition]

Nested List Comprehension [expression for item1 in iterable1 for item2 in iterable2]




```python
### Basic List Comprehension

square=[num**2 for num in range(10)]
print(square)
```

```python
## List comprehension with condition

lst3 = []
for i in range(20):
    if i%2 == 0:
        lst3.append(i)

print(lst3)
```

```python
evens = [x for x in range(20) if x%2 == 0]
print(evens)
```

```python
## Nested List Comphrension

lst1=[1,2,3,4]
lst2=['a','b','c','d']

pairs = [[i,j] for i in lst1 for j in lst2]
print(pairs)
```

```python
## List Comprehension with function calls

words = ["hello", "world", "python", "list", "comprehension"]
lengths = [len(word) for word in words]
print(lengths)
```

```python
# Real-world example: filtering active users

users = [
    {"name": "Saloni", "active": True},
    {"name": "Shizi", "active": False},
    {"name": "James", "active": True},
]

def get_active_users(users):
    active_users = [user for user in users if user["active"]]
    return active_users

print(get_active_users(users))
```

### Time Complexity Notes

- Access by index -> O(1)
- Append -> O(1) amortized
- Insert at beginning -> O(n)
- Remove -> O(n)
- Membership check (in) -> O(n)
