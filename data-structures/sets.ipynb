#### Sets
Sets are a built-in data type in Python used to store collections of unique items. They are unordered, meaning that the elements do not follow a specific order, and they do not allow duplicate elements. Sets are useful for membership tests, eliminating duplicate entries, and performing mathematical set operations like union, intersection, difference, and symmetric difference.

### When to Use Sets

Sets are useful when:
- You need **unique elements**
- Fast **membership testing**
- Performing **mathematical set operations**

Example use cases:
- Removing duplicates from data
- Finding common users between systems
- Fast lookups

```python
### create a set

my_set = {10, 20, 30, 10, 20, 40}
print(my_set)
print(type(my_set))
```

```python
my_empty_set = set()
print(my_empty_set)
print(type(my_empty_set))
```

```python
my_set1 = set([10, 20, 30, 10, 20, 40])
print(my_set1)
print(type(my_set1))
```

```python
## Basics Sets Operation
## Adiing and Removing Elements

my_set1.add(50)
print(my_set1)
my_set1.remove(20)
print(my_set1)
```

```python
my_set1.remove(100)
print(my_set1)
```

```python
my_set1.discard(110)
print(my_set1)
```

```python
## pop method
popped_item = my_set1.pop()
print(popped_item)
print(my_set1)
```

```python
## clear the set

my_set1.clear()
print(my_set1)
```

```python
## Set Memebership test

my_set2 = {11, 22, 33, 44, 55, 66, 20, 30, 40}
print(22 in my_set2)
print(10 in my_set2)
```

```python
## Mathematical Operation

my_set1 = {10, 20, 30, 40, 50, 22, 44, 33, 11}
union_set = my_set1.union(my_set2)
print(union_set)
```

```python
intersected_set = my_set1.intersection(my_set2)
print(intersected_set)
```

```python
my_set1.intersection_update(my_set2)
print(my_set1)
```

```python
## Difference

set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

difference_set = set1.difference(set2)
print(difference_set)
```

```python
## Sets Methods
set1={1,2,3,4,5}
set2={3,4,5}

## is subset
print(set1.issubset(set2))

print(set1.issuperset(set2))
```

```python
lst=[1,2,2,3,4,4,5]

set(lst)
```

```python
### Counting Unique words in text

text="In this tutorial we are discussing about sets"
words=text.split()

## convert list of words to set to get unique words

unique_words=set(words)
print(unique_words)
print(len(unique_words))
```

```python
# Example: finding common users between two platforms

platform_a = {"saloni", "rahul", "john", "emma"}
platform_b = {"emma", "rahul", "alex"}

common_users = platform_a.intersection(platform_b)

print(common_users)
```

### Time Complexity Notes

- Membership check → O(1)
- Add element → O(1)
- Remove element → O(1)
- Union / Intersection → O(n)

Sets are implemented using **hash tables**, which makes lookups very fast.

#### Conclusion
Sets are a powerful and flexible data type in Python that provide a way to store collections of unique elements. They support various operations such as union, intersection, difference, and symmetric difference, which are useful for mathematical computations. Understanding how to use sets and their associated methods can help you write more efficient and clean Python code, especially when dealing with unique collections and membership tests.
