##### Introduction to Dictionaries

Dictionaries are unordered collections of items. They store data in key-value pairs.
Keys must be unique and immutable (e.g., strings, numbers, or tuples), while values can be of any type.

### Time Complexity

- Access value by key → O(1)
- Insert key-value pair → O(1)
- Delete key → O(1)

- Dictionaries are implemented using hash tables, which makes key lookups very fast.

```python
## Creating Dictionaries

empty_dict = {}
print(type(empty_dict))
print(empty_dict)
```

```python
empty_dict = dict()
print(type(empty_dict))
```

```python
student_details = {
    "name": "Saloni",
    "age": 26,
    "city": "Bangalore",
    "profession": "Software Engineer",
    "grade": 14
}

print(student_details)
print(type(student_details))
```

```python
# Single key is always used to access the value in a dictionary

temp_dict = {
    "name" : "Alice",
    "age": 20,
    "name": "Bob"
}

print(temp_dict)
```

```python
## accessing Dictionary Elements

print(student_details)
print(student_details["profession"])
```

```python
## Accessing Dictionary elements
print(student_details["name"])
print(student_details["age"])
print(student_details["city"])
print(student_details["profession"])
print(student_details["grade"])
```

```python
## Accessing using get() method

print(student_details.get("name"))
print(student_details.get("age"))
print(student_details.get("city"))
print(student_details.get("profession"))
print(student_details.get("grade"))

print(student_details.get("job_type", "Not found"))
```

```python
## Modifying Dictionary Elements
## Dictionaries are mutable, so you can add, update, or delete elements

print(student_details)
```

```python
##update value for the key
student_details["age"] = 25
print(student_details)

## added a new key and value pair to the dictionary
student_details["hobby"] = "Singing"
print(student_details)
```

```python
del student_details["grade"]

print(student_details)
```

```python
## Dictionary methods

##get all the keys
keys = student_details.keys()
print(keys)

## get all the values
key_values = student_details.values()
print(key_values)

## get all the key-value pairs
items = student_details.items()
print(items)
```

```python
## Shallow copy of a dictionary

student_details_copy = student_details
print(student_details)
print(student_details_copy)
```

```python
student_details["name"] = "Shizi"
print(student_details)
print(student_details_copy)
```

```python
student_deets_copy1 = student_details.copy()
print(student_deets_copy1)
print(student_details)
```

```python
student_details["grade"] = 15
print(student_details)
print(student_deets_copy1)
```

```python
### Iterating Over Dictionaries
## You can use loops to iterate over dictionaries, keys, values, or items

## Iterating over keys
for key in student_details.keys():
    print(key)
```

```python
## Iterating over values

for value in student_details.values():
    print(value)
```

```python
## Iterate over key value pairs

for key, value in student_details.items():
    print(f"{key}: {value}")
```

```python
## Nested Dictionaries


students={
    "student1":{"name":"Krish","age":32},
    "student2":{"name":"Peter","age":35}
}
print(students)
```

```python
## Access nested dictionaries elements

print(students["student2"]["name"])
print(students["student2"]["age"])
```

```python
students.items()
```

```python
## Iterating over nested dictionaries

for student_key, student_info in students.items():
    print(f"{student_key}: {student_info}")
    for key, value in student_info.items():
        print(f"  {key}: {value}")
```

```python
## Dictionary Comprehension

squares={x:x**2 for x in range(5)}
print(squares)
```

```python
## Condition dictionary comprehension

evens={x:x**2 for x in range(10) if x%2==0}
print(evens)
```

```python
## Practical Examples

## USe a dictionary to count the frequency of elements in list

numbers=[1,2,2,3,3,3,4,4,4,4]
frequency={}

for number in numbers:
    if number in frequency:
        frequency[number]+=1
    else:
        frequency[number]=1
print(frequency)
```

```python
## Merge 2 dictionaries into one

dict1={"a":1,"b":2}
dict2={"b":3,"c":4}
merged_dict={**dict1,**dict2}
print(merged_dict)
```

```python
# Example: counting words in a sentence

text = "python is powerful and python is easy"

word_freq = {}

for word in text.split():
    word_freq[word] = word_freq.get(word, 0) + 1

print(word_freq)
```

#### Conclusion
Dictionaries are powerful tools in Python for managing key-value pairs. They are used in a variety of real-world scenarios, such as counting word frequency, grouping data, storing configuration settings, managing phonebooks, tracking inventory, and caching results. Understanding how to leverage dictionaries effectively can greatly enhance the efficiency and readability of your code.
