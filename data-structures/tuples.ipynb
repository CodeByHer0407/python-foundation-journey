##### Introduction to Tuples
Explanation:

Tuples are ordered collections of items that are immutable.
They are similar to lists, but their immutability makes them different.


### When to Use Tuples

Use tuples when:
- Data should not change
- Returning multiple values from a function
- Using fixed structured data

### Performance Note

Tuples are slightly faster than lists because they are immutable.
They also consume less memory.

```python
## creating a tuple

empty_tuple = ()
print(empty_tuple)
print(type(empty_tuple))
```

```python
lst = list()
print(type(lst))
my_tup = tuple()
print(type(my_tup))
```

```python
single_element = (5,)
print(type(single_element))
```

```python
numbers = tuple([10, 20, 30, 40, 50])
print(numbers)
print(type(numbers))
```

```python
list((11, 22, 33, 44, 55, 66))
```

```python
mixed_tuple = (1, "Hello", 3.14, True)
print(mixed_tuple)
```

```python
## Accessing Tuple Elements

numbers
```

```python
print(numbers[2])
print(numbers[-1])
```

```python
print(numbers[0:4])
```

```python
numbers[::-1]
```

```python
## Tuple Operations

concatenated_tuple = numbers + mixed_tuple
print(concatenated_tuple)
```

```python
mixed_tuple * 3
```

```python
## Immutable Nature Of Tuples
## Tuples are immutable, meaning their elements cannot be changed once assigned.

lst=[1,2,3,4,5]
print(lst)

lst[1]="Krish"
print(lst)
```

```python
numbers[2] = 100
print(numbers)
```

```python
## Tuple Methods

print(numbers.count(30))
print(numbers.index(30))
```

```python
## Packing and Unpacking tuple
## packing

my_packed_tuple = 1, "Hello", 3.14, True
print(my_packed_tuple)
```

```python
##unpacking a tuple

a, b, c, d = my_packed_tuple
print(a)
print(b)
print(c)
print(d)
```

```python
## Unpacking with *

packed_tuples = (11, 20, 30, 40, 50, 60, 70, 80, 99)
first, *middle, last = packed_tuples
print(first)
print(middle)
print(last)
```

```python
## Nested Tuple
## Nested List
lst=[[1,2,3,4],[6,7,8,9],[1,"Hello",3.14,"c"]]
lst[0][0:3]
```

```python
lst=[[1,2,3,4],[6,7,8,9],(1,"Hello",3.14,"c")]
lst[2][0:3]
```

```python
nested_tuple = ((1, 2, 3), ("a", "b", "c"), (True, False))

## access the elements inside a tuple
print(nested_tuple[0])
print(nested_tuple[1][2])
```

```python
## iterating over nested tuples

for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item,end=" ")
    print()
```

```python
## Example of packing and unpacking in a function

def get_coordinates():
    return (12.9716, 77.5946)

lat, lon = get_coordinates()

print(lat, lon)
```
